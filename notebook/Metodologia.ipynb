```python
import pandas as pd
import glob

# 1. Cargar el dataset principal (Board_Summary_Cleaned)
df_summary = pd.read_csv("Board_Summary_Cleaned.csv")
df_summary.columns = df_summary.columns.str.strip()  # Eliminar espacios extra

# 2. Obtener la lista de archivos de compensaciÃ³n (por ejemplo, "compensation_1999.csv", etc.)
comp_files = sorted(glob.glob("Europe - SMDEs Compensation - *.csv"))

# 3. Definir las claves comunes para la fusiÃ³n
common_keys = ["DirectorID*", "CompanyID*", "Company Name", "Annual Report Year"]

# 4. Iterar sobre cada archivo de compensaciÃ³n y realizar la fusiÃ³n
for file in comp_files:
    # Cargar el archivo de compensaciÃ³n
    df_comp = pd.read_csv(file)
    df_comp.columns = df_comp.columns.str.strip()  # Limpiar nombres de columnas
    
    # Si en el archivo de compensaciÃ³n la columna aparece sin el asterisco, renombrarla:
    if "DirectorID" in df_comp.columns and "DirectorID*" not in df_comp.columns:
        df_comp.rename(columns={"DirectorID": "DirectorID*"}, inplace=True)
    
    # Seleccionar Ãºnicamente las columnas de interÃ©s
    cols_comp = [
        "DirectorID*", "CompanyID*", "Company Name", "Annual Report Year",
        "Total Compensation", "Equity Linked", "Total Wealth"
    ]
    
    # Verificar que todas las columnas existan en df_comp
    missing_cols = [col for col in cols_comp if col not in df_comp.columns]
    if missing_cols:
        print(f"El archivo {file} no tiene las columnas: {missing_cols}. SaltÃ¡ndolo.")
        continue  # O puedes decidir otra acciÃ³n segÃºn el caso
    
    df_comp_reduced = df_comp[cols_comp].copy()
    
    # Renombrar columnas de compensaciÃ³n para que coincidan con Board Summary - 1:
    df_comp_reduced.rename(columns={
        "Total Compensation": "Total Annual Compensation",
        "Equity Linked": "Total Equity Linked Compensation"
    }, inplace=True)
    
    # Obtener el aÃ±o (suponiendo que todos los registros del archivo corresponden al mismo aÃ±o)
    year = df_comp_reduced["Annual Report Year"].iloc[0]
    
    # Renombrar las columnas de compensaciÃ³n aÃ±adiendo un sufijo del aÃ±o para diferenciarlas
    df_comp_reduced.rename(columns={
        "Total Annual Compensation": f"Total Annual Compensation_{year}",
        "Total Equity Linked Compensation": f"Total Equity Linked Compensation_{year}",
        "Total Wealth": f"Total Wealth_{year}"
    }, inplace=True)
    
    # Realizar el LEFT JOIN sobre las claves comunes
    df_summary = df_summary.merge(df_comp_reduced, on=common_keys, how="left")
    print(f"FusiÃ³n realizada para el aÃ±o {year}")

# Guardar el dataset final fusionado en un archivo CSV
df_summary.to_csv("Board_Summary_Fusionado.csv", index=False)
```

```python
import pandas as pd
import numpy as np
from sklearn.impute import KNNImputer

# Cargar el dataset
df_fusionado = pd.read_csv("Board_Summary_Fusionado.csv")
df_fusionado.columns = df_fusionado.columns.str.strip()

ml_columns = [
    "Nationality Mix",
    "Time to Retirement",
    "Age (Yrs)",
    "Avg. Yrs on Other Quoted Boards",
    "Total Number of Quoted Boards to Date"
]

# Seleccionar un subconjunto de 1000 filas para la prueba
df_subset = df_fusionado.sample(10000, random_state=42)

# Convertir las columnas a numÃ©rico
for col in ml_columns:
    df_subset[col] = pd.to_numeric(df_subset[col], errors="coerce")

print("Valores nulos en el subconjunto de 1000 filas:")
print(df_subset[ml_columns].isnull().sum())

# Inicializar KNNImputer con 5 vecinos
knn_imputer = KNNImputer(n_neighbors=5)
transformed = knn_imputer.fit_transform(df_subset[ml_columns])
print("Forma de la matriz transformada:", transformed.shape)

# Si la forma transformada es (1000, 5), entonces puedes asignar sin problemas:
if transformed.shape[1] == len(ml_columns):
    df_subset[ml_columns] = transformed
    print("ImputaciÃ³n con KNN completada en el subconjunto:")
    print(df_subset[ml_columns].head())
else:
    print("La matriz transformada tiene una forma inesperada. Revisa los datos de entrada.")
```

```python
import pandas as pd
import numpy as np
from sklearn.impute import KNNImputer

# 1. Ajustar el imputador en una muestra representativa
# Cargamos, por ejemplo, las primeras 5000 filas para obtener un sample
sample_df = pd.read_csv("Board_Summary_Fusionado.csv", nrows=5000)
sample_df.columns = sample_df.columns.str.strip()  # Limpiar nombres de columnas

# Definir las columnas a imputar (ajusta segÃºn tus necesidades)
ml_columns = [
    "Time to Retirement",
    "Age (Yrs)",
    "Avg. Yrs on Other Quoted Boards",
    "Total Number of Quoted Boards to Date"
]

# Convertir estas columnas a numÃ©rico (cualquier error se convertirÃ¡ en NaN)
for col in ml_columns:
    sample_df[col] = pd.to_numeric(sample_df[col], errors="coerce")

# Inicializar y ajustar el KNNImputer con 5 vecinos (ajusta n_neighbors si es necesario)
knn_imputer = KNNImputer(n_neighbors=5)
knn_imputer.fit(sample_df[ml_columns])
print("Imputador KNN ajustado en el sample.")

# 2. Procesar el dataset completo en chunks para no cargar todo en memoria
chunksize = 10000  # Ajusta el tamaÃ±o del chunk segÃºn tu memoria
chunks = pd.read_csv("Board_Summary_Fusionado.csv", chunksize=chunksize)
imputed_chunks = []

# Procesar cada chunk individualmente
for chunk in chunks:
    # Limpiar nombres de columnas
    chunk.columns = chunk.columns.str.strip()
    
    # Convertir las columnas de interÃ©s a numÃ©rico
    for col in ml_columns:
        chunk[col] = pd.to_numeric(chunk[col], errors="coerce")
    
    # Aplicar el imputador ya ajustado a las columnas seleccionadas
    chunk[ml_columns] = knn_imputer.transform(chunk[ml_columns])
    
    # Almacenar el chunk imputado
    imputed_chunks.append(chunk)
    print("Procesado un chunk de tamaÃ±o:", len(chunk))

# 3. Concatenar todos los chunks imputados
df_imputed = pd.concat(imputed_chunks, ignore_index=True)

# 4. Guardar el dataset final imputado en un archivo CSV
df_imputed.to_csv("KNN_Board_Summary_Filtered.csv", index=False)
print("KNN Imputation completada con chunks y guardada en 'KNN_Board_Summary_Filtered.csv'.")
```

```python
import pandas as pd
import numpy as np

# Cargar el dataset (ajusta la ruta/nombre del archivo segÃºn corresponda)
df = pd.read_csv("Board_Summary_Fusionado.csv", low_memory=False)
df.columns = df.columns.str.strip()  # Limpiar espacios en los nombres de columnas

# Definir las columnas a imputar con la media
columns_to_mean = ["Time on Board", "Time in Role", "Time in Company", "Number of Qualifications"]

# Convertir a numÃ©rico y aplicar imputaciÃ³n por la media
for col in columns_to_mean:
    # Convertir la columna a tipo numÃ©rico; los errores se convierten en NaN
    df[col] = pd.to_numeric(df[col], errors="coerce")
    
    # Calcular la media (excluyendo nulos)
    mean_val = df[col].mean()
    
    # Imputar los valores nulos con la media
    df[col] = df[col].fillna(mean_val)
    
    print(f"Columna '{col}' imputada con la media: {mean_val}")

# Guardar el DataFrame resultante en un nuevo archivo CSV
output_file = "Board_Summary_MeanImputed.csv"
df.to_csv(output_file, index=False)
print(f"ImputaciÃ³n con la media completada y guardada en '{output_file}'.")
```
